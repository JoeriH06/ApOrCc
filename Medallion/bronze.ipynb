In this bronze layer we will gather the data from the api and load it into a dataframe.  
First we import all necessary libraries for the bronze layer.

```python
import pandas as pd
import os
from dotenv import load_dotenv

```

Following this up we start with getting the API request

```python
# Setting API URL and parameters
load_dotenv()
API_PASSWORD = os.getenv("ENAPPSYS_API_PASSWORD")
API_USER = os.getenv("ENAPPSYS_API_USER")

url = f"https://app.enappsys.com/datadownload?code=nl/elec/pricing/daprices&currency=EUR&delimiter=semicolon&minavmax=false&pass={API_PASSWORD}&res=hourly&tag=csv&timezone=CET&user={API_USER}&start=202602101425&end=202602112220"
```

Now we check if that worked and got a correct call

```python
try:
    df_bronze = pd.read_csv(url, delimiter=';')
    print("Data loaded successfully.")
except Exception as e:
    print(f"An error occurred while loading the data: {e}")

print(df_bronze.head())
```

```python
# save the bronze data in folder data/bronze
df_bronze.to_csv('../Data/bronze.csv', index=False)
```
